```python
import pandas as pd  # data munching and data wrangling
import mlxtend  # python lib for data science tasks 
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
import pandas as pd 
df= pd.read_csv(r"C:\Users\halge\Downloads\Retail_Data.csv")
```

```python
df.head()
```

```python
# making a truth table
df1 = pd.get_dummies(df)
df2 = df1.iloc[:,1:].astype(int)
df2.head()  # if the customer has purchased then true if not false 
```

```python
frequent_items = apriori(df2,min_support = 0.05 , use_colnames = True)
frequent_items
```

```python
rules = association_rules(frequent_items, metric ="lift", min_threshold = 1)
type(rules) 
rules.shape
```
